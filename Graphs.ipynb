```python
df1['smoker'].value_counts()
```

```python
sns.relplot(data = df1, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data = df1, x = 'total_bill', y = 'tip', hue='smoker')
# Here the "hue" is for group by the data
```

```python
sns.relplot(data = df1, x = 'total_bill', y = 'tip', style='smoker')
```

```python
sns.relplot(data = df1, x = 'total_bill', y = 'tip', style = 'size',color='red')
```

```python
sns.relplot(data = df1, x = 'total_bill', y = 'tip', style = 'time',hue = 'smoker',size = 'tip')
```

```python
sns.relplot(data=df1,x='total_bill',y='tip',col='day')
```

```python
sns.relplot(data=df1,x='total_bill',y='tip',col='time')
```

### Categorical_Plot

```python
sns.catplot(x = 'day', y = 'total_bill', data = df1)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = df1,kind = 'box')
```

```python
sns.catplot(data=df1,x='time',y='tip', kind = 'bar')
```

```python
sns.catplot(x = 'sex', y = 'tip', data = df1,kind = 'swarm')
```
