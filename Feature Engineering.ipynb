### Numerical feature Handling

```python
numerical_feature_nan = [features for features in df.columns if df[features].dtypes!='object' and df[features].isna().sum() >=1]
```

```python
numerical_feature_nan
```

```python
df[numerical_feature_nan].head(10)
```

```python
for features in numerical_feature_nan:
    
    # since there are outlier so we need to replace by median
    
    median_values = df[features].median()
    
    # create new feature so that capture nan value
    
    df[features + 'nan'] = np.where(df[features].isna(),1,0)
    df[features].fillna(median_values,inplace=True)
    
```

### Date and Time variable

```python
for features in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[features + 'age of house'] = df['YrSold'] - df[features]
```
