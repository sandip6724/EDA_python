#### 1. Missing values

```python
df.isna().sum()
```

```python
# Separate missing values
features_with_na = [features for features in df.columns if df[features].isna().sum() >=1]
print(len(features_with_na))
```

```python
features_with_na
```

```python
df[features_with_na]
```

```python
# percentage of missing values
for features in features_with_na:
    print(features,np.round(df[features].isna().mean()*100,4),'% missing value')
```

#### Lets plot some Graphs

```python
data = df.copy()
for features in features_with_na:
    # lets take 1 for missing value and 0 for present
    data[features] = np.where(data[features].isna(),1,0)
    # Lets find mean sales price where data is missing
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.ylabel('SalePrice')
    plt.show()
```

#### Note : In the above figures shows that the relation between the missing values and dependent variable is clearly visible. so we need to replace these nan values with something meaningful , which can be perform in the Feature Engineering

### 2. Numerical Variable

```python
df.info()
```

```python
numerical_features = [features for features in df.columns if df[features].dtypes!='object']
print(len(numerical_features))
```

```python
numerical_features
```

```python
df[numerical_features].head()
```

```python
# Temporal Variables ( Date and Time )
year_features = [features for features in numerical_features if 'Yr' in features or 'Year' in features]
```

```python
year_features
```

```python
for features in year_features:
    print(features,df[features].unique())
```

```python
df.groupby('YearBuilt')['SalePrice'].median().plot.bar()
plt.xlabel('YearBuilt')
plt.ylabel('Median SalePrice')
plt.title(features)
plt.show()
```

```python
df.groupby('YearRemodAdd')['SalePrice'].median().plot.bar()
plt.xlabel('YearRemodAdd')
plt.ylabel('Median SalePrice')
plt.title(features)
plt.show()
```

```python
df.groupby('YrSold')['SalePrice'].median().plot.bar()
plt.xlabel('YrSold')
plt.ylabel('median SalePrice')
plt.title(features)
plt.show()
```

```python
# compare the difference between all the year features with sales price
data = df.copy()
for features in year_features:
    if features != 'YrSold':
        data[features] = data['YrSold'] - data[features]
        plt.scatter(data[features],data['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('House sales price')
        plt.show()
```

#### Note : With all the year wise info , it is clearly visible that the new houses are costiler than that of older one.

### Numerical variables are 2 types
1. Discrete variable
2. contineous variable

```python
# 1. Discrete
discrete_features = [features for features in numerical_features if len(df[features].unique())<=25]
```

```python
discrete_features
```

```python
df[discrete_features].head()
```

```python
# Lets find the relationship between discrete features with sales price
data = df.copy()
for features in discrete_features:
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.xlabel(features)
    plt.ylabel('Sales price')
    plt.show()
```

#### Note : There is a relationship between discrete and sales price

```python
# 2. contineous
contineous_features = [features for features in numerical_features if features not in discrete_features + year_features]
print(len(contineous_features))
```

```python
contineous_features
```

```python
df[contineous_features].head()
```

```python
# find the relation between contineous features with sales price
data = df.copy()
for features in contineous_features:
    data.groupby(features)['SalePrice'].median().plot.hist()
    plt.xlabel(features)
    plt.ylabel('sales price')
    plt.title(features)
    plt.show()
```

```python
# Lets analyze the contineous values by creating histogram plot to understand the distribution
data = df.copy()
for features in contineous_features:
    data[features].hist(bins=25)
    plt.xlabel(features)
    plt.ylabel('frequency')
    plt.show()
```

```python
# logarithmic transformation
data=df.copy()
for features in contineous_features:
    if 0 in data[features].unique():
        pass
    else:
        data[features]=np.log(data[features])
        plt.scatter(data[features],data['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('SalesPrice')
        plt.title(features)
        plt.show()
```

### 5. Outlier

```python
%matplotlib inline
data = df.copy()
for features in contineous_features:
    if 0 in data[features].unique():
        pass
    else:
        data[features]=np.log(data[features])
        data.boxplot(column=features)
        plt.ylabel(features)
        plt.show()
```
